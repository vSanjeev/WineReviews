```python
%matplotlib inline
import pandas as pd
```

```python
wine = pd.read_csv('data/wine-reviews/winemag-data-130k-v2.csv', low_memory=False)
```

```python
del wine['Unnamed: 0']
```

## Category Variables
1. Country
2. Designation?
3. Province
4. Region 1
5. Region 2
6. Variety
7. Winery

```python
wine.dtypes
```

```python
wine.country = wine.country.astype('category')
wine.designation = wine.designation.astype('category')
wine.province = wine.province.astype('category')
wine.region_1 = wine.region_1.astype('category')
wine.region_2 = wine.region_2.astype('category')
wine.variety = wine.variety.astype('category')
wine.winery = wine.winery.astype('category')
```

```python
wine.dtypes
```

```python
wine.description[:-10]
```

```python
wine.country.value_counts()[:10].plot(kind='barh', rot=0)
```

```python
t_wine_US = wine[wine.country=='US']
t_wine_US.province.value_counts()[:10].plot(kind='barh', rot=0)
```

```python
t_wine_US_Cal = t_wine_US[t_wine_US.province=='California']
t_wine_US_Cal.region_1.value_counts()[:10].plot(kind='barh', rot=10)
```
